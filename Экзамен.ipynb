```
class Tag:
  def __init__(self):
    self.name=""
    self.text=""
    self.atr=""
    self.box=[]

  def generate(self):
    txt=""
    txt="<"+self.name+" "+self.atr+">"+self.text+" \n"
    for i in range(len(self.box)):
      txt=txt+self.box[i].generate()
    txt=txt+"<"+self.name+"> \n"
    return txt

class Img(Tag):
  def __init__(self, a, t):
    Tag.__init__(self)
    self.name="img"
    self.atr=a
    self.text=t

class Html(Tag):
  def __init__(self, b):
    Tag.__init__(self)
    self.name="html"
    self.box=b

class Head(Tag):
  def __init__(self, a, b):
    Tag.__init__(self)
    self.name="head"
    self.atr=a
    self.box=b

class Title(Tag):
  def __init__(self, t):
    Tag.__init__(self)
    self.name="title"
    self.text=t

class Meta(Tag):
  def __init__(self, a, t):
    Tag.__init__(self)
    self.name="meta"
    self.atr=a
    self.text=t

class Body(Tag):
  def __init__(self, a, t, b):
    Tag.__init__(self)
    self.name="body"
    self.atr=a
    self.text=t
    self.box=b

class Prf(Tag):
  def __init__(self, a, t ,b):
    Tag.__init__(self)
    self.name="p"
    self.atr=a
    self.text=t
    self.box=b

class H1(Tag):
  def __init__(self, a, t, b):
    Tag.__init__(self)
    self.name="h1"
    self.atr=a
    self.text=t

class H2(Tag):
  def __init__(self, a, t, b):
    Tag.__init__(self)
    self.name="h2"
    self.atr=a
    self.text=t

class H3(Tag):
  def __init__(self, a, t, b):
    Tag.__init__(self)
    self.name="h3"
    self.atr=a
    self.text=t

class H4(Tag):
  def __init__(self, a, t, b):
    Tag.__init__(self)
    self.name="h4"
    self.atr=a
    self.text=t

class H5(Tag):
  def __init__(self, a, t, b):
    Tag.__init__(self)
    self.name="h5"
    self.atr=a
    self.text=t

class H6(Tag):
  def __init__(self, a, t, b):
    Tag.__init__(self)
    self.name="h6"
    self.atr=a
    self.text=t


a=Html([Head("", [Meta("", 'charset="utf-8"'), Title("Example test")]), Body("", "", [H1("", "Exam work"), Prf("", "This text is example my exam work", []), H4("", "Warning"), Prf("", "I hope this work is not very boring. But just in case, keep the picture.", []), Img('src="https://img4.goodfon.ru/original/2080x1269/d/53/krolik-listik-listia-listva-osen.jpg"', 'alt="Example image"')])])
b=a.generate()
file = open("Exp.html", "w")
file.write(b)
file.close()

    
```
